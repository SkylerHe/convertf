```python
#dfall1.ind.txt was used
#create a new file
popfile_samplename = open('/scratch/myang_shared/lab/Skyler/popfile_samplename.txt','w')

#load ind file
indfile = open('/scratch/myang_shared/lab/Skyler/zind_prune_i_100_10_2_dfall1_pop.ind','r')
indline = indfile.readlines()

#get the unique populaiton names
poplist = []
for x in indline:
    indlines = x.split('\t')
    if indlines[2][:-1] not in poplist:
        poplist.append(indlines[2][:-1])
        

#list the populations that Dr.E collected
source1 = ['MIA_2019_June',
           'CM_2017_early',
           'CM_2017_late',
           'CM_2019_late',
           'CM_2018_early',
           'CM_2018_late',
           'CM_2019_mid',
           'HPO_2019_mid',
           'HPO_2020_early',
           'CM_2020_early']

#list the populations that are in the western hemisphere 
source2 = ['NC',
           'HI',
           'TN',
           'MIA_2019_June',
           'CM_2017_early',
           'CM_2017_late',
           'CM_2019_late',
           'CM_2018_early',
           'CM_2018_late',
           'CM_2019_mid',
           'HPO_2019_mid',
           'HPO_2020_early',
           'CM_2020_early']
#run two for loops to compare source1 and source2 except for themselves
for constant in source1:
    for pop in source2:
        if constant != pop:
            popfile_samplename.write(constant + '\t' + pop + '\t' + 'Kenya' + '\n')

#close files
popfile_samplename.close()
indfile.close()
```
